* Задание 2   
Проведите тест гипотезы. Утверждается, что шарики для подшипников, изготовленные автоматическим станком, имеют средний диаметр 17 мм. Используя односторонний критерий с α=0,05, проверить эту гипотезу, если в выборке из n=100 шариков средний диаметр оказался равным 17.5 мм, а дисперсия известна и равна 4 кв. мм.

```python
xav = 17.5
mu = 17
sig = 2
n = 100
z = (xav - mu) / (sig / n**0.5)
print(f'В соответствии с таблицей z-значений, точка отсечки альфа = 1,645')
if z < 1.645:
  print(f'Мы получили значение z равным {z}, поскольку оно меньше 1,645, делаем вывод, что наше значение попадает в область принятия H0, то есть достоверных отклонений в размерах шариков нет')
else:
  print(f'Мы получили значение z равным {z}, поскольку оно больше 1,645, делаем вывод, что наше значение попадает в область принятия H1, то есть выявлены достоверные отклонения в размерах шариков')
```

* Задание 3   
Проведите тест гипотезы. Продавец утверждает, что средний вес пачки печенья составляет 200 г. Из партии извлечена выборка из 10 пачек. Вес каждой пачки составляет: 202, 203, 199, 197, 195, 201, 200, 204, 194, 190. Известно, что их веса распределены нормально. Верно ли утверждение продавца, если учитывать, что доверительная вероятность равна 99%? (Провести двусторонний тест.)

```python
def mean(lst):
  return sum(lst) / len(lst)

def unbaised_dispersion(lst):
  res = 0
  for i in lst:
    res += (i-mean(lst))**2
  return res/(len(lst)-1)
  
```

```python
pacsweight = [202, 203, 199, 197, 195, 201, 200, 204, 194, 190]
n = len(pacsweight)
mu = sum(pacsweight)/n
xav = 200
sig_n = unbaised_dispersion(pacsweight)**0.5
t = (xav - mu) / (sig_n / n**0.5)
print(f'sig_n = {sig_n};  t = {t}')
print(f'По таблице значений Стьюдента, t для P = 0,995 для степени свободы 9 равен 3,2498')
print(f'Расчётное значение t = {t}, следовательно, утверждение продавца верно, поскольку t  расчётное меньше t табличного')
```

* Задание 4   
Есть ли статистически значимые различия в росте дочерей? Рост матерей 172, 177, 158, 170, 178,175, 164, 160, 169, 165 Рост взрослых дочерей: 173, 175, 162, 174, 175, 168, 155, 170, 160

```python
import numpy as np

height_moth = np.array([172, 177, 158, 170, 178, 175, 164, 160, 169, 165])
height_daug = np.array([173, 175, 162, 174, 175, 168, 155, 170, 160])
m_mean = np.mean(height_moth)
d_mean = np.mean(height_daug)
disp_m = np.var(height_moth, ddof=1)
disp_d = np.var(height_daug, ddof=1)
nm, nd = len(height_moth), len(height_daug)
crit_Stud = 1.7396
t = (m_mean - d_mean) / ((disp_m/nm) + (disp_d/nd))
if t < crit_Stud:
  print(f't = {t}, что меньше t табличного для 95% ДИ, следовательно статистически значимых различий нет')
else:
  print(f't = {t}, что ,больше t табличного для 95% ДИ, следовательно статистически значимые различия есть')
```
